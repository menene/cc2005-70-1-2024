# Pokemon

```python
import pandas as pd
```

Abrir el csv con datos del pokemon

```python
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon
```

```python
poderosos = pokemon.loc[pokemon["Attack"] >= 100]

print(poderosos)
```

```python
mitad = pokemon.loc[pokemon["#"] < 400]

print(mitad)
```

```python
legendarios_hielo = pokemon.loc[(pokemon["Legendary"] == True) & (pokemon["Type 1"] == "Ice")]

print(legendarios_hielo[["#", "Name", "Type 1", "Type 2", "Attack", "Defense"]])
```

```python
dos_a_cuatro = pokemon.loc[(pokemon["#"] >= 200) & (pokemon["#"] <= 400)]

print(dos_a_cuatro)
```

```python
poderosos_legendarios = pokemon.loc[(pokemon["Attack"] >= 100) & (pokemon["Legendary"] == True)]

print(poderosos_legendarios["Name"])
```

Es posible usar la funciÃ³n query para hacer mÃ¡s "natural" la bÃºsqueda

```python
poderosos_fantasmas = pokemon.query("Attack >= 10 and `Type 1` == 'Ghost'")

print(poderosos_fantasmas)
```

Se pueden utilizar ciclos para recorrer la informaciÃ³n

```python
for i in range(len(pokemon)):
    if (pokemon.loc[i]["Legendary"] and pokemon.loc[i]["Type 1"] == "Fire"):
        print(pokemon.loc[i]["Name"] + " - " + str(pokemon.loc[i]["Attack"]) + "/" + str(pokemon.loc[i]["Defense"]))
```
